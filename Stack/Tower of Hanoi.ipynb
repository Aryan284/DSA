```python
def tower(n, from_rod, aux_rod, to_rod):
    if n == 1:
        print("Move disk 1", "from", from_rod, "to", to_rod)
    else:
        tower(n-1, from_rod, to_rod, aux_rod)
        print("Move disk",n,"from", from_rod, "to", to_rod)
        tower(n-1, aux_rod, from_rod, to_rod)
n = int(input("Enter number of disk: "))
tower(n, 'A', 'B', 'C')


2nd method--

def toh(n, from_rod, to_rod, aux_rod):
    if n == 1:
        print("Move disk 1", "from", from_rod, "to", to_rod)
    else:
        toh(n-1, from_rod, aux_rod, to_rod)
        print("Move disk",n,"from", from_rod, "to", to_rod)
        toh(n-1, aux_rod, to_rod, from_rod)
n = int(input("Enter number of disk: "))
toh(n, 'A', 'C', 'B')

Iterative Solution---



class Node():
    def __init__(self, val):
        self.data = val
        self.next = None
        
class Stack():
    def __init__(self):
        self.head = None

    def push(self, stack, res):
        temp = Node(res)
        temp.next = stack.head
        stack.head = temp
        return stack

    def pop(self, stack):
        if not stack.head:
            return -1, stack
        popped = stack.head.data
        ptr = stack.head
        stack.head = ptr.next
        ptr = None
        return popped, stack

    def top(self,stack):
         if not stack.head:
                print("Stack is empty!")
                return stack
        return stack.head.data
         

    def isempty(self, stack):
        if not stack.head:
            return True
        else:
            return False

    def display(self, stack):
        if not stack.head:
            print("Stack is Empty!")
            return stack
        temp = stack.head
        while temp is not None:
            print(temp.data, end = "->")
            temp = temp.next
        print("X")



    def print_moves(self, s, d, n):
        print("Move disk", n, 'from', s, 'to', d)

    def toh(self, n, source, aux, dest):
        s = 'S'
        a = 'A'
        d = 'D'

        if n % 2 == 0:
            a, d = d, a

        no_moves = pow(2, n)-1

        for i in range(n, 0, -1):
            source = source.push(source, i)
        for i in range(0, no_moves):
            if i % 3 == 0:
                source, dest = source.move_disk(source, dest, s, d)
            elif i % 3 == 1:
                source, aux = source.move_disk(source, aux, s, a)
            elif i % 3 == 2:
                aux, dest = source.move_disk(aux, dest, a, d)

    def move_disk(self, source, dest, s, d):
        popped_element1, source = source.pop(source)
        popped_element2, dest = dest.pop(dest)
        if popped_element1 == -1:
            source = source.push(source, popped_element2)
            source.print_moves(d, s, popped_element2)

        elif popped_element2 == -1:
            dest = dest.push(dest, popped_element1)
            dest.print_moves(s, d, popped_element1)

        elif popped_element1 > popped_element2:
            source = source.push(source, popped_element1)
            source = source.push(source, popped_element2)
            source.print_moves(d, s, popped_element2)

        else:
            dest = dest.push(dest, popped_element2)
            dest = dest.push(dest, popped_element1)
            dest.print_moves(s, d, popped_element1)
        return source, dest

if __name__=='__main__':
    stack = Stack()
    source = Stack()
    aux = Stack()
    dest = Stack()
    while True:
        print("-1 to break")
        print("1.Tower of Hanoi(Iterative)")
        ch = int(input("Enter your choice: "))
        if ch == -1:
            break
        elif ch == 1:
            n = int(input("Enter the number of disk: "))
            stack.toh(n, source, aux, dest)
```
