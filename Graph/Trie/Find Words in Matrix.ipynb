```python
# Using Dictionary
class Trie():
    def __init__(self):
        self.children = {}
        self.isend = False
        
class Solution:
    def findWords(self, board: List[List[str]], words: List[str]) -> List[str]:
        m = len(board)
        n = len(board[0])
        root = Trie()
        
        for i in words:
            self.insert(root, i)
        res = set()
        visited = [[False for _ in range(n)] for _ in range(m)]
        
        for i in range(m):
            for j in range(n):
                ch = board[i][j]
                if ch in root.children:
                    res = self.helper(root.children[ch], board, i, j, m, n, ch, res, visited)
        return res
    
    def helper(self, root, board, i, j, m, n, path, res, visited):
        x = [0,-1,0,1]
        y = [-1,0,1,0]
        
        if root.isend:
            res.add(path)
        visited[i][j] = True
        
        for key, value in root.children.items():
            for k in range(4):
                if i + x[k] >= 0 and i + x[k] < m and j + y[k] >= 0 and j + y[k] < n:
                    if board[i + x[k]][j + y[k]] == key and not visited[i + x[k]][j + y[k]]:
                        self.helper(value, board, i + x[k], j + y[k], m, n, path + key, res, visited)
                            
        visited[i][j] = False
        return res
    
    def insert(self, root, string):
        ptr = root
        for i in (string):
            ptr = ptr.children.setdefault(i, Trie())
        ptr.isend = True
```
