### Linked list to BST

```python
class linked_list():
    def __init__(self, val):
        self.data = val
        self.next = None
class bst():
    def __init__(self ,val):
        self.data = val
        self.left = None
        self.right = None
        
def make_list(elements):
    head = linked_list(elements[0])
    for element in elements[1:]:
        ptr = head
        while ptr.next:
            ptr = ptr.next
        ptr.next = linked_list(element)
    return head

def printbst(root):
    if root is not None:
        print(root.data)
        printbst(root.left)
        printbst(root.right)
        
def convertbst(node):
    if not node:
        return None
    if not node.next:
        return bst(node.data)
    slow = fast = node
    prev = None
    while fast and fast.next:
        prev = slow
        slow = slow.next
        fast = fast.next.next
        
    prev.next = None
    root = bst(slow.data)
    root.left = convertbst(node)
    root.right = convertbst(slow.next)
    return root
    
head = make_list([-10, -3, 0, 5, 9])
root = convertbst(head)
printbst(root)
```

### Sorted Array to BST 

```python
class bst():
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
def convertbst(arr):
    if not arr:
        return
    mid = len(arr)//2
    root = bst(arr[mid])
    root.left = convertbst(arr[:mid])
    root.right = convertbst(arr[mid+1:])
    return root
def preorder(root):
    if not root:
        return
    print(root.data)
    preorder(root.left)
    preorder(root.right)

if __name__=="__main__":
    arr = [1, 2, 3, 4, 5, 6, 7]
    root = convertbst(arr)
    print("\nPreorder traversal")
    preorder(root)
```

### Greater Sum Tree

```python
class Bst():
    def __init__(self, val):
        self.data = val
        self.left = None
        self.right = None
        
def insert(root, val):
    if root is None:
        return Bst(val)
    if root.data < val:
        root.right = insert(root.right, val)
    else:
        root.left = insert(root.left, val)
    return root

    
def inorder(root):
    if not root:
        return None
    inorder(root.left)
    print(root.data)
    inorder(root.right)
    
def transform(root):
    sum = 0
    def greatersum(root):
        nonlocal sum
        if not root:
            return 
        greatersum(root.right)
        sum = sum + root.data
        root.data = sum - root.data
        greatersum(root.left)
    greatersum(root)
    return sum
    
if __name__=="__main__":
    root = Bst(11)
    root = insert(root, 2)
    root = insert(root, 29)
    root = insert(root, 1)
    root = insert(root, 7)
    root = insert(root, 15)
    root = insert(root, 40)
    root = insert(root, 35)
    transform(root)
    print("\n\n")
    inorder(root)
```
