```python
class Node:
    def __init__(self, val):
        self.data = val
        self.next = None
        
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
 
    def istop(self, queue):
        if not queue.front:
            print("Queue is empty!")
        else:
            return queue.front.data
 
    def isempty(self, queue):
        if not queue.front:
            return True
        else:
            return False
 
    def create(self, queue):
        if queue.front:
            print("Queue already created!")
            return queue
        while True:
            print("-1 to exit")
            n = int(input("Enter number want to added: "))
            if n == -1:
                break
            newnode = Node(n)
            if queue.isempty(queue):
                queue.front = newnode
                queue.rear = newnode
            else:
                queue.rear.next = newnode
                queue.rear = newnode
        return queue
    def display(self, queue):
        if queue.isempty(queue):
            print("Queue is Empty!")
            return queue
        temp = queue.front
        while temp is not None:
            print(temp.data, end = "|")
            temp = temp.next
        print("X")
 
    def enqueue(self, queue):
        while True:
            print("-1 to exit")
            n = int(input("Enter number to enqueue: "))
            if n == -1:
                break
            newnode = Node(n)
            if queue.isempty(queue):
                queue.front = newnode
                queue.rear = newnode
            else:
                queue.rear.next = newnode
                queue.rear = newnode
        return queue
        
    def dequeue(self, queue):
        if queue.isempty(queue):
            print("Queue is empty!")
            return queue
        temp = queue.front
        remove = queue.front.data
        queue.front = temp.next
        temp = None
        return remove, queue
                    
 
 
 
 
 
 
if __name__=='__main__':
    queue = Queue()
    while True:
        print("-1 to break")
        print("1. Create Queue")
        print("2. Display")
        print("3. Enqueue")
        print("4. Dequeue")
        print("5.Top element")
        ch = int(input("Enter your choice: "))
        if ch == 1:
            queue = queue.create(queue)
        elif ch == 2:
            queue.display(queue)
        elif ch == 3:
            queue = queue.enqueue(queue)
        elif ch == 4:
            remove, queue = queue.dequeue(queue)
            print("Remove element is", remove)
        elif ch == 5:
            front = queue.istop(queue)
            print("Front element is", front)
 
        else:
            break
```
