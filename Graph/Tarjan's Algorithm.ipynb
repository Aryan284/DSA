```python
from collections import defaultdict

class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
        self.scc_count = 0
        self.max = 0
    
    def add_edge(self, u, v):
        self.graph[u].append(v)
#         self.graph[v].append(u)
        self.max = max(u, v, self.max)
    
    def tarjan_algo(self):
        def dfs(u):
            stack.append(u)
            on_stack[u] = True
            ids[u] = low[u] = self.node_id
            self.node_id += 1
            for v in self.graph[u]:
                if ids[v] == UNVISITED:
                    dfs(v)
                if on_stack[v]:
                    low[u] = min(low[u], low[v])
            if ids[u] == low[u]:
                while stack:
                    node = stack.pop()
                    print(node, end = ' ')
                    on_stack[node] = False
                    low[node] = ids[u]
                    if node == u:
                        break
                self.scc_count += 1
                print()
        
        UNVISITED, self.node_id = -1, 0
        n = self.max + 1
        ids, low, on_stack, stack = [UNVISITED] * n, [float('inf')] * n, [False] * n, []
        for i in range(n):
            if ids[i] == UNVISITED:
                dfs(i)
        return low

if __name__=='__main__':
    g1 = Graph()
    g1.add_edge(1, 0)
    g1.add_edge(0, 2)
    g1.add_edge(2, 1)
    g1.add_edge(0, 3)
    g1.add_edge(3, 4)
    print("nSSC in first graph ")
    low = g1.tarjan_algo()
#     print(low)
#     print(g1.scc_count)
    g2 = Graph()
    g2.add_edge(0, 1)
    g2.add_edge(1, 2)
    g2.add_edge(2, 3)
    print("nSSC in second graph ")
    g2.tarjan_algo()


    g3 = Graph()
    g3.add_edge(0, 1)
    g3.add_edge(1, 2)
    g3.add_edge(2, 0)
    g3.add_edge(1, 3)
    g3.add_edge(1, 4)
    g3.add_edge(1, 6)
    g3.add_edge(3, 5)
    g3.add_edge(4, 5)
    print("nSSC in third graph ")
    g3.tarjan_algo()

    g4 = Graph()
    g4.add_edge(0, 1)
    g4.add_edge(0, 3)
    g4.add_edge(1, 2)
    g4.add_edge(1, 4)
    g4.add_edge(2, 0)
    g4.add_edge(2, 6)
    g4.add_edge(3, 2)
    g4.add_edge(4, 5)
    g4.add_edge(4, 6)
    g4.add_edge(5, 6)
    g4.add_edge(5, 7)
    g4.add_edge(5, 8)
    g4.add_edge(5, 9)
    g4.add_edge(6, 4)
    g4.add_edge(7, 9)
    g4.add_edge(8, 9)
    g4.add_edge(9, 8)
    print("nSSC in fourth graph ")
    g4.tarjan_algo()


    g5 = Graph()
    g5.add_edge(0, 1)
    g5.add_edge(1, 2)
    g5.add_edge(2, 3)
    g5.add_edge(2, 4)
    g5.add_edge(3, 0)
    g5.add_edge(4, 2)
    print("nSSC in fifth graph ")
    g5.tarjan_algo()
    
    g6 = Graph()
    g6.add_edge(1, 3)
    g6.add_edge(0, 1)
    g6.add_edge(1, 2)
    g6.add_edge(2, 0)
    print("nSSC in fifth graph ")
    g6.tarjan_algo()
```
