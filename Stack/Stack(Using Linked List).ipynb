```python
class Node():
    def __init__(self, val):
        self.data = val
        self.next = None
class Stack():
    def __init__(self):
        self.head = None
    def create(self, stack):
        if stack.head:
            print("Stack is already Created!")
            return stack
        while True:
            print("-1 to exit")
            n = int(input("Enter the number want to push: "))
            if n == -1:
                break
            temp = Node(n)
            if stack.head is None:
                stack.head = temp
            else:
                temp.next = stack.head
                stack.head = temp
        return stack
    def display(self, stack):
        if not stack.head:
            print("Stack is Empty!")
            return stack
        temp = stack.head
        while temp is not None:
            print(temp.data, end = "->")
            temp = temp.next
        print("X")
    def push(self, stack):
        n =int(input("Enter the number want to push: "))
        temp = Node(n)
        if not stack.head:
            stack.head = temp
            return stack
        else:
            temp.next = stack.head
            stack.head = temp
            return stack
    def pop(self, stack):
        if not stack.head:
            print("No element to pop!")
            return stack
        popped = stack.head.data
        ptr = stack.head
        stack.head = ptr.next
        ptr = None
        return popped, stack
    def top(self,stack):
        if not stack.head:
            print("Stack is empty!")
            return stack
        top = stack.head.data
        return top
    def isempty(self, stack):
        if not stack.head:
            print("Stack is empty!")
        else:
            print("Stack is not empty!")




if __name__=='__main__':
    stack = Stack()
    while True:
        print("-1 tot break")
        print("1.Create a stack")
        print("2. Display a stack")
        print("3. Push in stack")
        print("4. Pop in stack")
        print("5. Top in stack")
        print("6. Empty function")
        ch = int(input("Enter your choice: "))

        if ch == 1:
            stack = stack.create(stack)
        elif ch == 2:
             stack.display(stack)
        elif ch == 3:
            stack = stack.push(stack)
        elif ch == 4:
            popped, stack = stack.pop(stack)
            print("Popped element is", popped)
        elif ch == 5:
            top = stack.top(stack)
            print("Top element is", top)
        elif ch == 6:
            stack.isempty(stack)
        else:
            break
        
```
