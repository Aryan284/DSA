```python
def submatrix(i, j):
    if i < 3 and j < 3:
        return 0
    if i < 3 and j < 6:
        return 1
    if i < 3 and j < 9:
        return 2
    if i < 6 and j < 3:
        return 3
    if i < 6 and j < 6:
        return 4
    if i < 6 and j < 9:
        return 5
    if i < 9 and j < 3:
        return 6
    if i < 9 and j < 6:
        return 7
    return 8

def solve(arr):
    row = [[0 for _ in range(10)] for _ in range(10)]
    col = [[0 for _ in range(10)] for _ in range(10)]
    cell = [[0 for _ in range(10)] for _ in range(10)]
    for i in range(len(arr)):
        for j in range(len(arr[0])):
            num = arr[i][j]
            if num != 0:
                row[i][num] = 1
                col[j][num] = 1
                cell[submatrix(i, j)][arr[i][j]] = 1
    return row, col, cell
def valid_sudoku(k, row, col, cell, i, j):
    if row[i][k] == 1:
        return False
    if col[j][k] == 1:
        return False
    if cell[submatrix(i, j)][k] == 1:
        return False
    return True

def sudoku_solve(arr, row, col, cell):
    count = 0
    for i in range(9):
        for j in range(9):
            if arr[i][j] == 0:
                count = 1
    if count == 0:
        return True
    for i in range(9):
        for j in range(9):
            if arr[i][j] == 0:
                flag = True
                for k in range(1, 10):
                    if valid_sudoku(k, row, col, cell, i, j):
                        flag = False
                        arr[i][j] = k
                        row[i][arr[i][j]] = 1
                        col[j][k] = 1
                        cell[submatrix(i,j)][k] = 1
                        if sudoku_solve(arr, row, col, cell):
                            return True
                        arr[i][j] = 0
                        row[i][k] = 0
                        col[j][k] = 0
                        cell[submatrix(i,j)][k] = 0
                if flag or arr[i][j] == 0:
                    return False
    return False


arr = [[5,3,0,0,7,0,0,0,0],
        [6,0,0,1,9,5,0,0,0],
        [0,9,8,0,0,0,0,6,0],
        [8,0,0,0,6,0,0,0,3],
        [4,0,0,8,0,3,0,0,1],
        [7,0,0,0,2,0,0,0,6],
        [0,6,0,0,0,0,2,8,0],
        [0,0,0,4,1,9,0,0,5],
        [0,0,0,0,8,0,0,7,9]]
row, col, cell = solve(arr)
sudoku_solve(arr, row, col, cell)
for i in arr:
    print(i)
```
