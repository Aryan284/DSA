```python
class Node:
    def __init__(self, val):
        self.prev = None
        self.data = val
        self.next = None

class Deque:
    def __init__(self):
        self.front = None
        self.rear = None
    def isEmpty(self, deque):
        if deque.front is None:
            return True
        else:
            return False
    def create(self, deque):
        if not deque.isEmpty(deque):
            print("Already created!")
            return deque
        while True:
            print("-1 to break")
            n = int(input("Enter number want to add: "))
            if n == -1:
                break
            temp = Node(n)
            if deque.isEmpty(deque):
                deque.front = temp
                deque.rear = temp
            else:
                deque.rear.next = temp
                temp.prev = deque.rear
                deque.rear = temp
        return deque

    def display(self, deque):
        if deque.isEmpty(deque):
            print("Deque is Empty!")
        temp = deque.front
        while temp is not None:
            print(temp.data, end = "->")
            temp = temp.next
        print("X")

    def insertfront(self, deque):
        n = int(input("Enter number added: "))
        temp = Node(n)
        if deque.isEmpty(deque):
            deque.front = temp
            deque.rear = temp
        else:
            temp.next = deque.front
            deque.front.prev = temp
            deque.front = temp
        return deque

    def insertend(self, deque):
        n = int(input("Enter number added: "))
        temp = Node(n)
        if deque.isEmpty(deque):
            deque.front = temp
            deque.rear = temp
        else:
            temp.prev = deque.rear.next
            deque.rear.next = temp
            deque.rear = temp  
        return deque

    def deletefront(self, deque):
        if deque.isEmpty(deque):
            print("Empty delete not possible")
            return deque
        if deque.front.next is None:
            deque.front = None
            deque.rear = None
            return deque
        temp = deque.front
        deque.front = deque.front.next
        deque.front.prev = None
        temp = None
        return deque
    
    def deleteend(self, deque):
        if deque.isEmpty(deque):
            print("Empty delete not possible")
            return deque
        if deque.front.next is None:
            deque.front = None
            deque.rear = None
            return deque
        temp = deque.rear
        deque.rear = deque.rear.prev
        deque.rear.next = None
        temp = None
        return deque

    def getrear(self, deque):
        if deque.isEmpty(deque):
            print("Deque Empty")
            return
        print("Rear value is", deque.rear.data)
    
    def getfront(self, deque):
        if deque.isEmpty(deque):
            print("Deque Empty")
            return deque
        print("Front value is", deque.front.data)
        


if __name__=="__main__":
    deque = Deque()
    while True:
        print("-1 to break")
        print("1. Create")
        print("2. Display")
        print("3. Insertfront")
        print("4. Insertend")
        print("5. Deletefront")
        print("6. Deleteend")
        print("7. Deque Front")
        print("8. Deque Rear")
        ch = int(input("Enter your choice: "))
        if ch == 1:
            deque = deque.create(deque)
        elif ch == 2:
            deque.display(deque)
        elif ch == 3:
            deque = deque.insertfront(deque)
        elif ch == 4:
            deque = deque.insertend(deque)
        elif ch == 5:
            deque = deque.deletefront(deque)
        elif ch == 6:
            deque = deque.deleteend(deque)
        elif ch == 7:
            deque.getfront(deque)
        elif ch == 8:
            deque.getrear(deque)
        else:
            break
```
