```python
class Node():
    def __init__(self, val):
        self.data = val
        self.next = None
class Stack():
    def __init__(self):
        self.head = None

    def push(self, stack, res):
        temp = Node(res)
        temp.next = stack.head
        stack.head = temp
        return stack

    def pop(self, stack):
        if not stack.head:
            return -1, stack
        popped = stack.head.data
        ptr = stack.head
        stack.head = ptr.next
        ptr = None
        return popped, stack

    def top(self,stack):
        if not stack.head:
            print("Stack is empty!")
             return stack
        return stack.head.data
         

    def isempty(self, stack):
        if not stack.head:
            return True
        else:
            return False

    def create(self, stack):
        if stack.head:
            print("Stack is already Created!")
            return stack
        while True:
            print("-1 to exit")
            n = int(input("Enter the number want to push: "))
            if n == -1:
                break
            temp = Node(n)
            if stack.head is None:
                stack.head = temp
            else:
                temp.next = stack.head
                stack.head = temp
        return stack

    def display(self, stack):
        if not stack.head:
             print("Stack is Empty!")
             return stack
        temp = stack.head
        while temp is not None:
            print(temp.data, end = "->")
            temp = temp.next
        print("X")

    def prefix_eval(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        for  i in expression:
            if i.isdigit():
                stack = stack.push(stack, int(i))
            else:
                op1, stack = stack.pop(stack)
                op2, stack = stack.pop(stack)
                if i == "+":
                    res = int(op2) + int(op1)
                elif i == "*":
                    res = int(op2) * int(op1)
                elif i == "/":
                    res = int(op2) / int(op1)
                elif i == "-":
                    res = int(op2) - int(op1)
                stack = stack.push(stack, res)
        print("Required value is", stack.top(stack))
        return stack
    def postfix_eval(self, stack, expression):
        expression = expression.split()
        for i in expression:
            if i.isdigit():
                stack = stack.push(stack, int(i))
            else:
                op1, stack = stack.pop(stack)
                op2, stack = stack.pop(stack)
                if i == "+":
                    res = int(op2) + int(op1)
                elif i == "-":
                    res = int(op2) - int(op1)
                elif i == "*":
                    res = int(op2) * int(op1)
                elif i == "/":
                    res = int(int(op2) / int(op1))
                stack = stack.push(stack, res)
        print("Required value is", stack.top(stack))
        return stack
    
    def higher_precedence(self, op1, op2):
        if op1 == "+" or op1 == "-":
            first_weight = 1
        elif op1 == "*" or op1 == "/":
            first_weight = 2
        if op2 == "+" or op2 == "-":
            second_weight = 1
        elif op2 == "*" or op2 == "/":
             second_weight = 2
        if first_weight <= second_weight:
            return True
        else:
            return False

    def open_parenthesis(self, operator):
        if operator == "(" or operator == "{" or operator == "[":
            return True
        else:
            return False
    def close_parenthesis(self, operator):
        if operator == ")" or operator == "}" or operator == "]":
            return True
        else:
            return False


    def infix_postfix(self, stack, expression):
        expression = expression.split()
        postfix = ""
        for i in expression:
            if i.isalnum():
                postfix += i + " "
            else:
                while not stack.isempty(stack) and not stack.higher_precedence( i, stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isempty(stack):
             popped_element, stack = stack.pop(stack)
             postfix += popped_element + " "
        return postfix
    
    def infix_prefix(self, stack,expression):
        expression = expression.split()
        expression.reverse()
        prefix = ""
        for i in expression:
            if i.isalnum():
                prefix += i + " "
            else:
                while not stack.isempty(stack) and not stack.higher_precedence(i,stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isempty(stack):
            popped_element, stack = stack.pop(stack)
            prefix += popped_element + " "
        return prefix[::-1]

    
    def infix_postfix_braces(self, stack, expression):
        expression = expression.split()
        postfix = ""
        for i in expression:
            if i.isalnum():
                postfix += i + " "
            elif stack.open_parenthesis(i):
                stack = stack.push(stack, i)
            elif stack.close_parenthesis(i):
                while not stack.isempty(stack) and not stack.open_parenthesis(stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix += popped_element + " "
                _, stack = stack.pop(stack)
            else:
                while not stack.isempty(stack) and not stack.open_parenthesis(stack.top(stack)) and not stack.higher_precedence(i, stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    postfix += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isempty(stack):
             popped_element, stack = stack.pop(stack)
             postfix += popped_element + " "
        return postfix

    def infix_prefix_braces(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        prefix = ""
        for i in expression:
            if i.isalnum():
                prefix += i + " "
            elif stack.close_parenthesis(i):
                stack = stack.push(stack, i)
            elif stack.open_parenthesis(i):
                while not stack.isempty(stack) and not stack.close_parenthesis(stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix += popped_element + " "
                _, stack = stack.pop(stack)
            else:
                while not stack.isempty(stack) and not stack.close_parenthesis(stack.top(stack)) and not stack.higher_precendence(i, stack.top(stack)):
                    popped_element, stack = stack.pop(stack)
                    prefix += popped_element + " "
                stack = stack.push(stack, i)
        while not stack.isempty(stack):
            popped_element, stack = stack.pop(stack)
            prefix += popped_element +  " "
        return prefix[::-1]

    def postfix_infix(self, stack, expression):
        expression = expression.split()
        infix = ""
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                op2, stack = stack.pop(stack)
                op1, stack = stack.pop(stack)
                res = op1 + " " + i + " " + op2
                stack = stack.push(stack, res)
        print("Infix is", stack.top(stack))
        return 

    def prefix_infix(self, stack, expression):
        expression = expression.split()
        expression.reverse()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                op2, stack = stack.pop(stack)
                op1, stack = stack.pop(stack)
                res = op1 + " " + i + " " + op2
                stack = stack.push(stack, res)
        infix_val = stack.top(stack)
        infix_val = infix_val.split()
        infix_val.reverse()
        return ' '.join(infix_val)

    def prefix_postfix(self, stack,expression):
        expression = expression.split()
        expression.reverse()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                op1, stack = stack.pop(stack)
                op2, stack = stack.pop(stack)
                res = op1 + " " + op2 + " " + i
                stack = stack.push(stack, res)
        postfix_val = stack.top(stack)
        return postfix_val

    def postfix_prefix(self, stack, expression):
        expression = expression.split()
        for i in expression:
            if i.isalnum():
                stack = stack.push(stack, i)
            else:
                op1, stack = stack.pop(stack)
                op2, stack = stack.pop(stack)
                res = i + " " + op2 + " " + op1
                stack = stack.push(stack, res)
        prefix_val = stack.top(stack)
        return prefix_val

    def insert_bottom(self, stack, item):
        if stack.isempty(stack):
            stack = stack.push(stack, item)
            return stack
        popped_element, stack = stack.pop(stack)
        stack = stack.insert_bottom(stack, item)
        stack = stack.push(stack, popped_element)
        return stack

    def reverse(self, stack):
        if not stack.isempty(stack):
            popped_element, stack = stack.pop(stack)
            stack = stack.reverse(stack)
            stack = stack.insert_bottom(stack, popped_element)
        return stack




             


if __name__=='__main__':
    stack = Stack()
    while True:
        print("-1 to break")
        print("1. Prefix evaluation")
        print("2. Display stack")
        print("3. Postfix evaluation")
        print("4. Infix to Postfix evaluation")
        print("5. Infix to Prefix evaluation")
        print("6. Infix to Postfix braces handled evaluation")
        print("7. Infix to Prefix braces handled evaluation")
        print("8. Postfix to infix")
        print("9. Prefix to infix")
        print("10. Prefix to postfix")
        print("11. Postfix to prefix")
        print("12. Create stack")
        print("13. Reverse stack")
        ch = int(input("Enter your choice: "))
        if ch == 1:
            expression = input("Enter the expression: ")
            stack = stack.prefix_eval(stack, expression)
            
        elif ch == 2:
            stack.display(stack)
        elif ch == 3:
            expression = input("Enter the expression: ")
            stack = stack.postfix_eval(stack, expression)
        elif ch == 4:
            expression = input("Enter the expression: ")
            postfix = stack.infix_postfix(stack, expression)
            print("Postfix expression is", postfix)
        elif ch == 5:
            expression = input("Enter the expression: ")
            prefix = stack.infix_prefix(stack, expression)
            print("Prefix expression is", prefix)
        elif ch == 6:
            expression = input("Enter the expression: ")
            postfix = stack.infix_postfix_braces(stack, expression)
            print("Postfix expression is", postfix)
        elif ch == 7:
            expression = input("Enter the expression: ")
            prefix = stack.infix_prefix_braces(stack, expression)
            print("Prefix expression is", prefix)
        elif ch == 8:
            expression = input("Enter the expression: ")
            stack = stack.postfix_infix(stack, expression)
        elif ch == 9:
            expression = input("Enter the expression: ")
            infix_val = stack.prefix_infix(stack, expression)
            print("Infix expression is", infix_val)
        elif ch == 10:
            expression = input("Enter the expression: ")
            postfix_val = stack.prefix_postfix(stack, expression)
            print("Postfix expression is", postfix_val)
        elif ch == 11:
            expression = input("Enter the expression: ")
            prefix_val = stack.postfix_prefix(stack, expression)
            print("Prefix expression is", prefix_val)
        elif ch == 12:
            stack = stack.create(stack)
        elif ch == 13:
            print("\n\n")
            print("Stack Reversed go to display to view!")
            stack = stack.reverse(stack)
            print("\n\n")
        else:
            break
```
